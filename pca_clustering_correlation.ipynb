```python
import pandas as pd
import matplotlib.pyplot as plt
df_main = pd.read_csv("crypto_30_365.csv", index_col = 0 )
df_main.head()
```

```python
df_main.isna().sum().sum()
```

```python
df_cleaned = df_main.fillna(method = "ffill", axis = 1).fillna(method="bfill", axis = 1)
df_cleaned.head()
```

```python
df_num = df_cleaned.select_dtypes(include=["number"])
df_num.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_cleaned)
```

```python
df_scaled = pd.DataFrame(scaled_data, index = df_cleaned.index, columns = df_cleaned.columns)
df_scaled.info()
```

```python
df_scaled.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
pca_result = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(pca_result, index = df_scaled.index, columns = ["PC1","PC2","PC3"])
df_pca.head()
```

```python


# Scatter plot of PCA components
plt.figure(figsize=(10, 6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], color='blue', alpha=0.7)

# Annotate points with cryptocurrency names
for i, txt in enumerate(df_cleaned.index):
    plt.annotate(txt, (df_pca['PC1'][i], df_pca['PC2'][i]), fontsize=9, alpha=0.7)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Visualization of 30 Cryptocurrencies")
plt.grid(True)
plt.show()
```

```python

from mpl_toolkits.mplot3d import Axes3D


# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with PCA components
ax.scatter(df_pca['PC1'], df_pca['PC2'], df_pca['PC3'], color='blue', alpha=0.7)

# Annotate points with cryptocurrency names
for i, txt in enumerate(df_cleaned.index):
    ax.text(df_pca['PC1'][i], df_pca['PC2'][i], df_pca['PC3'][i], txt, fontsize=8, alpha=0.7)

ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3")
ax.set_title("3D PCA Visualization of 30 Cryptocurrencies")

plt.show()
```

```python
import numpy as np
explained_variance = np.cumsum(pca.explained_variance_ratio_)  
print(explained_variance)
```

```python
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
crypto_names = df_cleaned.index
pca_components = df_pca.iloc[:,0:3]

#applying KMeans
kmeans  = KMeans(n_clusters = 4, random_state = 42, n_init="auto")
kmeans_labels = kmeans.fit_predict(pca_components)
```

```python
#applying Hierarchical clustering using ward's method

linkage_matrix = linkage(pca_components, method="ward", metric="euclidean")
hierarchical_labels = fcluster(linkage_matrix, 4, criterion = "maxclust")-1
```

```python
#comparing quality metrics for both clustering models
kmeans_silhouette = silhouette_score(pca_components, kmeans_labels)
hierarchical_silhouette = silhouette_score(pca_components, hierarchical_labels)

kmeans_ch = calinski_harabasz_score(pca_components, kmeans_labels)
hierarchical_ch = calinski_harabasz_score(pca_components, hierarchical_labels)

kmeans_db = davies_bouldin_score(pca_components, kmeans_labels)
hierarchical_db = davies_bouldin_score(pca_components, hierarchical_labels)
```

```python
# Print comparison metrics
print("Clustering Quality Metrics (4 clusters):")
print("-" * 50)
print(f"K-means Silhouette Score: {kmeans_silhouette:.4f}")
print(f"Hierarchical Silhouette Score: {hierarchical_silhouette:.4f}")
print()
print(f"K-means Calinski-Harabasz Score: {kmeans_ch:.2f}")
print(f"Hierarchical Calinski-Harabasz Score: {hierarchical_ch:.2f}")
print()
print(f"K-means Davies-Bouldin Score: {kmeans_db:.4f}")
print(f"Hierarchical Davies-Bouldin Score: {hierarchical_db:.4f}")
print("-" * 50)
```

```python
fig = plt.figure(figsize=(15, 6))

# K-means plot
ax1 = fig.add_subplot(121, projection='3d')
colors = ['blue', 'red', 'green', 'purple']

for i in range(4):
    cluster_points = pca_components[kmeans_labels == i]
    ax1.scatter(
        cluster_points.iloc[:, 0],
        cluster_points.iloc[:, 1],
        cluster_points.iloc[:, 2],
        c=colors[i],
        label=f'Cluster {i+1}',
        alpha=0.7,
        s=80
    )

ax1.set_title('K-means Clustering (4 clusters)')
ax1.set_xlabel('PCA Component 1')
ax1.set_ylabel('PCA Component 2')
ax1.set_zlabel('PCA Component 3')
ax1.legend()

# Hierarchical plot
ax2 = fig.add_subplot(122, projection='3d')

for i in range(4):
    cluster_points = pca_components[hierarchical_labels == i]
    ax2.scatter(
        cluster_points.iloc[:, 0],
        cluster_points.iloc[:, 1],
        cluster_points.iloc[:, 2],
        c=colors[i],
        label=f'Cluster {i+1}',
        alpha=0.7,
        s=80
    )

ax2.set_title('Hierarchical Clustering (4 clusters)')
ax2.set_xlabel('PCA Component 1')
ax2.set_ylabel('PCA Component 2')
ax2.set_zlabel('PCA Component 3')
ax2.legend()

plt.tight_layout()
```

```python
kmeans_results = pd.DataFrame({
    'Cryptocurrency': crypto_names,
    'Cluster': kmeans_labels + 1  # Making clusters start from 1 for readability
})

hierarchical_results = pd.DataFrame({
    'Cryptocurrency': crypto_names,
    'Cluster': hierarchical_labels + 1 
})

# Print K-means clusters
print("\nK-means Clustering Results:")
print("=" * 30)
for cluster in range(1, 5):
    crypto_in_cluster = kmeans_results[kmeans_results['Cluster'] == cluster]['Cryptocurrency'].tolist()
    print(f"Cluster {cluster}: {','.join(crypto_in_cluster)}")

# Print Hierarchical clusters
print("\nHierarchical Clustering Results:")
print("=" * 30)
for cluster in range(1, 5):
    crypto_in_cluster = hierarchical_results[hierarchical_results['Cluster'] == cluster]['Cryptocurrency'].tolist()
    print(f"Cluster {cluster}: {', '.join(crypto_in_cluster)}")
```

```python
import seaborn as sns
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Cryptocurrency Correlation Heatmap")
plt.show()
```

```python
correlation_matrix.index.name = None
correlation_pairs = correlation_matrix.unstack().reset_index()
correlation_pairs.head()
```

```python
correlation_pairs.columns = ['Crypto1', 'Crypto2', 'Correlation']

# Remove self-correlations (where Crypto1 == Crypto2)
correlation_pairs = correlation_pairs[correlation_pairs['Crypto1'] != correlation_pairs['Crypto2']]

# Get the most positively and negatively correlated pairs
top_positive_correlations = correlation_pairs.sort_values(by='Correlation', ascending=False).head(8)
top_negative_correlations = correlation_pairs.sort_values(by='Correlation', ascending=True).head(8)
print("Top Positive Correlations:")
print(top_positive_correlations[::2])
print("\nTop Negative Correlations:")
print(top_negative_correlations[::2])
```

```python
from sklearn.metrics import pairwise_distances_argmin_min
# Get cluster centroids in PCA space
centroids = kmeans.cluster_centers_

# Find closest asset to each centroid in PCA space
closest_indices, _ = pairwise_distances_argmin_min(centroids, df_pca)

# Get asset names of best representatives
representatives = df_pca.iloc[closest_indices].index.values
cluster_labels = [f"Cluster {i+1}" for i in range(len(representatives))]

# Create dictionary of best representatives
best_representatives = dict(zip(cluster_labels, representatives))

print(best_representatives)
```

```python
import yfinance as yf
import pandas as pd

# Define the clusters
clusters = {
    "Cluster 1": ["ADA-USD", "AVAX-USD", "DAI-USD", "DOGE-USD", "DOT-USD", "FIL-USD", "HBAR-USD", "ICP-USD", 
                  "LDO-USD", "LEO-USD", "LINK-USD", "LTC-USD", "MATIC-USD", "OKB-USD", "SHIB-USD", "SOL-USD", 
                  "SUI-USD", "TON-USD", "TRX-USD", "TUSD-USD", "USDC-USD", "USDT-USD", "XLM-USD", "XMR-USD", "XRP-USD"],
    "Cluster 2": ["BTC-USD", "WBTC-USD"],
    "Cluster 3": ["ETH-USD"],
    "Cluster 4": ["BCH-USD", "BNB-USD"]
}

# Fetch market data for each cluster
market_data = []

for cluster, tickers in clusters.items():
    for ticker in tickers:
        try:
            crypto = yf.Ticker(ticker)
            info = crypto.info  # Fetch metadata
            
            market_data.append({
                "Cluster": cluster,
                "Asset": ticker,
                "Market_Cap": info.get("marketCap", "N/A"),
                "Volume": info.get("volume24Hr", "N/A"),
                "Price": info.get("regularMarketPrice", "N/A")
            })
        except Exception as e:
            print(f"Error fetching data for {ticker}: {e}")

# Convert to DataFrame
df_market = pd.DataFrame(market_data)

# Display the market data
print(df_market)
```

```python
highest_market_cap_assets = df_market.loc[df_market.groupby("Cluster")["Market_Cap"].idxmax()]
print("Assets with the highest Market Cap in each cluster:\n")
print(highest_market_cap_assets[["Cluster", "Asset", "Market_Cap"]])
```
